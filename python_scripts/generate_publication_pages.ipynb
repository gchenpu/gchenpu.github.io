````python
from pathlib import Path
import os

from publication_builder import PublicationGenerator

notebook_dir = Path.cwd()

generator = PublicationGenerator(
    bib_path="publications.bib",
    out_dir="../publication",
)
````

## Generate pages


````python
include_thesis = True
copy_thesis_pdf = True

count = generator.build_pages(
    include_thesis=include_thesis,
    copy_thesis_pdf=include_thesis and copy_thesis_pdf,
)

total = count + 1 if include_thesis else count
print(f"Generated {total} publication pages in '../publication/'")
if include_thesis and copy_thesis_pdf:
    print("Thesis page ensured in '../publication' and thesis.pdf synced to '../files'.")
````
