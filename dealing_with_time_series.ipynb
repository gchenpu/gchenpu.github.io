````python
import datetime
import numpy
import matplotlib.dates as md
import netCDF4
````

## Python's ```datetime```

https://docs.python.org/2/library/datetime.html

````python
datetime.datetime(2016, 4, 25) # (year, month, day, hour, minutes, seconds, microseconds) DDMMYYYY
````

````python
today = datetime.datetime(2016,4,25)
yesterday = datetime.datetime(2016,4,24)
````

````python
today - yesterday
````

````python
datetime.timedelta(days=5,hours=6)
````

````python
yesterday + datetime.timedelta(days=5,hours=6)
````

## NumPy's ```numpy.datetime64``` data type

http://docs.scipy.org/doc/numpy/reference/arrays.datetime.html

````python
numpy.datetime64('2005-02-25')
````

````python
numpy.datetime64('2005-02')
````

````python
numpy.datetime64('2005-02', 'D')
````

````python
numpy.datetime64('2005-02-01 00:00:00', 's')
````

````python
numpy.arange('2005-03', '2005-04', dtype='datetime64[D]')
````

## Matplotlib ```dates``` module

http://matplotlib.org/api/dates_api.html

* Gives dates with reference to the beginning of the Gregorian calendar, 0001-01-01 00:00:00.

````python
days = 365.25*2000
md.num2date(days)
````

## netCDF4 ```datetime``` module

https://netcdf4-python.googlecode.com/svn/trunk/docs/netCDF4-module.html

* VERY usefule when dealing with NetCDF files, which often have relative time units that are distinct from 0001-01-01.

````python
netCDF4.num2date(days, units='days since 1800-1-1 00:00:00', calendar='Gregorian')
````
